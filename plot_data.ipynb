```python
import pandas as pd

# Load the sampled datasets
cola_sample = pd.read_csv('cola_sample.csv')
sst2_sample = pd.read_csv('sst2_sample.csv')
mrpc_sample = pd.read_csv('mrpc_sample.csv')
mnli_sample = pd.read_csv('mnli_sample.csv')

# Verify the loaded datasets
print(cola_sample.head())
print(sst2_sample.head())
print(mrpc_sample.head())
print(mnli_sample.head())
```

```python
print("Sampled CoLA Class Distribution:", cola_sample['label'].value_counts())
print("Sampled SST-2 Class Distribution:", sst2_sample['label'].value_counts())
print("Sampled MRPC Class Distribution:", mrpc_sample['label'].value_counts())
print("Sampled MNLI Class Distribution:", mnli_sample['label'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot class distributions
plt.figure(figsize=(10, 6))
sns.countplot(x=cola_sample['label'])
plt.title('Balanced Class Distribution in CoLA Sample')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Unacceptable', 'Acceptable'])
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x=sst2_sample['label'])
plt.title('Balanced Class Distribution in SST-2 Sample')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Negative', 'Positive'])
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x=mrpc_sample['label'])
plt.title('Balanced Class Distribution in MRPC Sample')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Not Paraphrase', 'Paraphrase'])
plt.show()

mnli_labels = pd.Series(mnli_sample['label']).map({0: 'Entailment', 1: 'Neutral', 2: 'Contradiction'})
plt.figure(figsize=(10, 6))
sns.countplot(x=mnli_labels)
plt.title('Balanced Class Distribution in MNLI Sample')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
